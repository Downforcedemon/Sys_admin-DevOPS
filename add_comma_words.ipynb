```python
# Input numbers as a multiline string
numbers = """
$ add numbers here

"""  # Replace this with your numbers


# Convert the multiline string into a list of numbers, stripping whitespace
number_list = [line.strip() for line in numbers.splitlines() if line.strip()]

# Format each number with 'AS' in front, 'AM' at the back, and enclose in single quotes
formatted_numbers = [f"'AS{num}AM'" for num in number_list]

# Join the formatted numbers with a comma
result = ",".join(formatted_numbers)

# Print the result
print(result)
```
